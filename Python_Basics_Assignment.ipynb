Question 2 : Explain the difference between List and Tuple in Python?

A=>
| List                                | Tuple                                       |
| ----------------------------------- | ------------------------------------------- |
| A list is mutable (can be changed). | A tuple is immutable (cannot be changed).   |
| Uses square brackets `[ ]`.         | Uses parentheses `( )`.                     |
| Elements can be added or removed.   | Elements cannot be modified after creation. |
| Example: `[1, 2, 3]`                | Example: `(1, 2, 3)`                        |

Question 5 : Write a Python program to check whether a number is even or odd using conditional
statements.

```python
def evenorodd(number):
  if number % 2 == 0:
    return "Even"
  else:
    return "Odd"


print(evenorodd(51))
```

Question 7 : Create a list of five numbers and print the maximum number from the list.

```python
# creating a list of five numbers
numbers = [10,58,63,24,65,87,94,96,45,75,30]

# finding the maximum number
maximum_number = max(numbers)

# printing the maximum number
print("The maximum number is:", maximum_number)
```

Question 8 : Write a Python program to remove duplicate values from a list using a set.

```python
numbers = [10,30,20,10,30,50,40,60,40,80,20,40,60]
unique_numbers = list(set(numbers))
print(unique_numbers)
```
